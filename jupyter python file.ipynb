```python
# task 2 (read file and handle all null file inside nan)
import pandas as pd
df = pd.read_csv('orders.csv' , na_values = {'Not Available' ,'unknown' })
df["Ship Mode"].unique()
```

```python
# Task 2 - For good practice , we'll change the name of columns into lower case and separated by underscore
df.columns
df.rename(columns = {"Order Id": 'order_id' , "Ship Mode": 'ship_mode'}) # but here we've several columns for good practice we'll use an another approach
```

```python
# 2nd way is: convert list of columns to str and perform str functions
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" " , "_")
```

```python
df.head(5)
```

```python
# task 4 : derive new columns based on existing ones
df['discount_per_quantity'] = df['list_price'] * (df['discount_percent'] / 100)
df.head(3)
```

```python
# sale price column creation
df["sale_price_per_quantity"] = df["list_price"] - df["discount_per_quantity"]
df.head(3)
```

```python
# profit column creation
df["profit_per_quantity"] = df["sale_price_per_quantity"] - df["cost_price"]
df.head(3)
```

```python
# Change data type of order_date (wrong format cleaning)
df["order_date"] = pd.to_datetime(df["order_date"] , format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
# Drop Some non-required columns in origial data frame (inplace = true)
df.drop(columns = ["list_price" , "discount_percent" , "cost_price"] , inplace = True)
```
